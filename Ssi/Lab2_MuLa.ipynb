```python
import MuLambda

import random
import numpy as np
import matplotlib.pyplot as plt
from MuLambda import mu_lambda  # Importowanie funkcji z pliku MuLambda.py

# Parametry dla pierwszego przypadku
zadane_iteracje = [1, 3, 13]
mu = 4
lambd = 3
turniej_rozmiar = 2
mutacja_poziom = 3
iteracje_liczba = 20

# Uruchomienie algorytmu dla pierwszego przypadku
r, p = mu_lambda(mu, lambd, turniej_rozmiar, mutacja_poziom, iteracje_liczba, zadane_iteracje)

# Generowanie siatki i tła wykresu
x = np.linspace(1, 100, 100)
y = np.linspace(1, 100, 100)
X, Y = np.meshgrid(x, y)
tab = np.sin(Y * 0.05) + np.sin(X * 0.05) + 0.4 * np.sin(Y * 0.15) * np.sin(X * 0.15)

# Wykres 1 - Położenie punktów z puli rodzicielskiej przed ewolucją (po wylosowaniu)
plt.figure(figsize=(6, 6))
plt.contour(X, Y, tab, cmap='viridis', levels=10)
plt.scatter([ind[0] for ind in r[0]], [ind[1] for ind in r[0]], color='blue', label='Rodzice przed ewolucją')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.title('Położenie punktów przed ewolucją (Iteracja 0)')
plt.show()

# Wykres 2 - Położenie punktów z puli rodzicielskiej po 3 iteracjach ewolucji
plt.figure(figsize=(6, 6))
plt.contour(X, Y, tab, cmap='viridis', levels=10)
plt.scatter([ind[0] for ind in r[1]], [ind[1] for ind in r[1]], color='green', label='Rodzice po 3 iteracjach')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.title('Położenie punktów po 3 iteracjach (Iteracja 3)')
plt.show()

# Wykres 3 - Położenie punktów z puli rodzicielskiej po 13 iteracjach ewolucji
plt.figure(figsize=(6, 6))
plt.contour(X, Y, tab, cmap='viridis', levels=10)
plt.scatter([ind[0] for ind in r[2]], [ind[1] for ind in r[2]], color='red', label='Rodzice po 13 iteracjach')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.title('Położenie punktów po 13 iteracjach (Iteracja 13)')
plt.show()

# Parametry dla drugiego przypadku
turniej_rozmiar = 1
lambd = 1

# Uruchomienie algorytmu dla drugiego przypadku
r2, p2 = mu_lambda(mu, lambd, turniej_rozmiar, mutacja_poziom, iteracje_liczba, zadane_iteracje)

# Wykres 4 - Położenie punktów z puli rodzicielskiej przed ewolucją (po wylosowaniu) dla drugiego przypadku
plt.figure(figsize=(6, 6))
plt.contour(X, Y, tab, cmap='viridis', levels=10)
plt.scatter([ind[0] for ind in r2[0]], [ind[1] for ind in r2[0]], color='blue', label='Rodzice przed ewolucją')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.title('Położenie punktów przed ewolucją (Iteracja 0) - Przypadek 2')
plt.show()

# Wykres 5 - Położenie punktów z puli rodzicielskiej po 3 iteracjach ewolucji dla drugiego przypadku
plt.figure(figsize=(6, 6))
plt.contour(X, Y, tab, cmap='viridis', levels=10)
plt.scatter([ind[0] for ind in r2[1]], [ind[1] for ind in r2[1]], color='green', label='Rodzice po 3 iteracjach')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.title('Położenie punktów po 3 iteracjach (Iteracja 3) - Przypadek 2')
plt.show()

# Wykres 6 - Położenie punktów z puli rodzicielskiej po 13 iteracjach ewolucji dla drugiego przypadku
plt.figure(figsize=(6, 6))
plt.contour(X, Y, tab, cmap='viridis', levels=10)
plt.scatter([ind[0] for ind in r2[2]], [ind[1] for ind in r2[2]], color='red', label='Rodzice po 13 iteracjach')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.title('Położenie punktów po 13 iteracjach (Iteracja 13) - Przypadek 2')
plt.show()
```
